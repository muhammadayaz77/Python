```python
import pandas as pd
```

```python
dataset = pd.read_csv("Data-Files/HousePricesData.csv")
dataset
```

```python
x_area = dataset["House Size (sq meters)"].tolist()
print(x_area)
```

```python
y_price = dataset["Price (USD)"].tolist()
print(y_price)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_area,y_price,marker='x',color='r')
plt.title("House Sizes With Prices")
plt.xlabel("House Size in square meters(Ayaz)")
plt.ylabel("House Price in USD $")
plt.show()
```

```python
import random
w = round(random.random(), 2)
b = round(random.random(), 2)
print(w, b)
```

```python
def calculate_total_cost(x,y,w,b):
    m = len(y)
    total_cost = 0
    for i in range(m):
        y_predicted = w*x[i] + b
        cost = (y_predicted - y[i])**2
        total_cost += cost
    total_cost /= (2*m)
    return total_cost
```

```python
calculate_total_cost(x_area,y_price,w,b)
```
